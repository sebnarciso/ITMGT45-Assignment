```python
locations = {
    "admu":{
        "name":"Ateneo de Manila University",
        "latitude":14.6394557,
        "longitude":121.0758975        
    },
    "upd":{
        "name":"University of the Philippines-Diliman Campus",
        "latitude":14.6538519,
        "longitude":121.0663348
    },
    "dlsu":{
        "name":"De La Salle University-Taft",
        "latitude":14.5647346,
        "longitude":120.9909876

    },
}

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

initial_location=input("Enter source:")
end_destination=input("Enter destination:")

def eta(initial_location,end_destination):

    # Starting with UPD
    if initial_location == 'upd':
        if end_destination == 'admu':
            travel_time = legs[initial_location, end_destination]["travel_time_mins"]
        elif end_destination == 'dlsu':
            stop = 'admu'
            travel_time = legs[initial_location, stop]["travel_time_mins"] + legs[stop, end_destination]["travel_time_mins"]
        elif end_destination == 'upd':
            stop1 = 'admu'
            stop2 = 'dlsu'
            travel_time = legs[initial_location, stop1]["travel_time_mins"] + legs[stop1, stop2]["travel_time_mins"] + legs[stop2, end_destination]["travel_time_mins"]
   
    # Starting with ADMU
    elif initial_location == 'admu':
        if end_destination == 'dlsu':
            travel_time = legs[initial_location, end_destination]["travel_time_mins"]
        elif end_destination == 'upd':
            stop = 'dlsu'
            travel_time = legs[initial_location, stop]["travel_time_mins"] + legs[stop, end_destination]["travel_time_mins"]
        elif end_destination == 'admu':
            stop1 = 'dlsu'
            stop2 = 'upd'
            travel_time = legs[initial_location, stop1]["travel_time_mins"] + legs[stop1, stop2]["travel_time_mins"] + legs[stop2, end_destination]["travel_time_mins"]

    # Starting with DLSU
    elif initial_location == 'dlsu':
        if end_destination == 'upd':
            travel_time = legs[initial_location, end_destination]["travel_time_mins"]
        elif end_destination == 'admu':
            stop = 'upd'
            travel_time = legs[initial_location, stop]["travel_time_mins"] + legs[stop, end_destination]["travel_time_mins"]
        elif end_destination == 'dlsu':
            stop1 = 'upd'
            stop2 = 'admu'
            travel_time = legs[initial_location, stop1]["travel_time_mins"] + legs[stop1, stop2]["travel_time_mins"] + legs[stop2, end_destination]["travel_time_mins"]
   
    return travel_time
                
print("ETA:", eta(initial_location, end_destination))
```
